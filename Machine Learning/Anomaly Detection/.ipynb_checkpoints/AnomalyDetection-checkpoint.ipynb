## Import Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from adtk.data import validate_series
from adtk.visualization import plot
from adtk.detector import *
plt.style.use('ggplot')
```

## Data Preprocessing

```python
df = pd.read_csv('monthly')
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df = df['Mean']
```

```python
df.head(14)
```

```python
df.info()
```

###  ThresholdAD

```python
threshold_detector  = ThresholdAD(low=-0.3,high=0.5)
anomalies = threshold_detector.detect(df)
plot(df,anomaly=anomalies,anomaly_color='blue',anomaly_tag='marker',figsize=(10,8))

plt.show()
```

###  QuantileAD

```python
quantile_detector  = QuantileAD(low=0.01,high=0.9)
anomalies = quantile_detector.fit_detect(df)
plot(df,anomaly=anomalies,anomaly_color='blue',anomaly_tag='marker',figsize=(10,8))

plt.show()
```

### ESDT 

```python
df = validate_series(df)
esd_ad = GeneralizedESDTestAD(alpha=0.25)
anomalies = esd_ad.fit_detect(df)
plot(df,anomaly=anomalies,anomaly_color='blue',anomaly_tag='marker',figsize=(10,8))

plt.show()
```

### VolatilityShift 

```python
df = validate_series(df)
volatility_detector = VolatilityShiftAD(c=0.6,side='positive',window=10)
anomalies = volatility_detector.fit_detect(df)
plot(df,anomaly=anomalies,anomaly_color='blue',anomaly_tag='marker',figsize=(10,8))

plt.show()
```
